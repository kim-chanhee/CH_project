```python
import re

# Sample list of strings
strings = [
    "이니시스_0 그린고라운드-",
    "이니시스_0 러프스튜디-",
    "이니시스_0 엠스트-",
    "이니시스_0-",
    "이니시스_1 대운상사-",
    "이니시스_1 주식회사티와-"
]

# Regular expression pattern to match the desired words
pattern = re.compile(r'이니시스_\d+ (\S+)-')

# Extracting words that match the pattern
extracted_words = [pattern.search(s).group(1) for s in strings if pattern.search(s)]

print(extracted_words)
```

이니시스_: 문자열이 "이니시스_"로 시작해야 한다는 것을 의미합니다.

\d+: 숫자가 하나 이상 이어지는 패턴입니다. \d는 숫자(0-9)를 의미하며, +는 하나 이상을 의미합니다. 예를 들어, "0", "1", "123" 등이 가능합니다.

(\S+): 이 부분이 좀 더 중요합니다. 공백(space) 문자가 하나 있어야 하고, 그 뒤에 " (\S+)"이 있습니다.

\S+: 공백 문자가 아닌 모든 문자가 하나 이상 이어지는 패턴입니다. \S는 공백이 아닌 문자를 의미하며, +는 하나 이상을 의미합니다. 여기서 중요한 것은 단어 사이의 공백을 포함하지 않는다는 것입니다.

()는 그룹을 형성합니다. 정규 표현식에서 괄호는 매칭된 부분을 그룹으로 저장하여 나중에 참조할 수 있게 합니다.

-: 문자열이 "-"로 끝나야 한다는 것을 의미합니다.

예제 텍스트가 "이니시스_0 그린고라운드-"인 경우를 생각해봅시다:

1. 정규표현식 \b는 단어 경계를 인식합니다.

2. [가-힣]+는 한글 문자가 하나 이상 연속된 경우를 찾습니다. 즉, "그린고라운드" 같은 한글 단어를 찾습니다.

3. 마지막으로 다시 \b가 단어의 끝을 인식합니다.

예제 텍스트에서 "이니시스_0 그린고라운드-"를 정규표현식으로 처리하면 "그린고라운드"라는 한글 단어만 추출됩니다.

```python
import pandas as pd
import re

# 예시 데이터프레임 생성
data = {'inis': ['이니시스_0 그린고라운드-', '이니시스_(주)위메프 위메프-', '이니시스 인증연동코리아트라이브 유한', '이니시스(일반)-꽃도매119-', '이니시스_YES24-']}
df = pd.DataFrame(data)

# 제외할 문자 및 패턴 정의
exclude_patterns = r'이니시스|\(.*?\)|'

# 단어 추출 함수 정의
def extract_word(text):
    text = re.sub(exclude_patterns, '', text)  # 제외할 문자 제거
    words = re.findall(r'\b[가-힣A-Za-z0-9]+\b', text)  # 한글, 영어 및 숫자가 포함된 단어 추출
    return ' '.join(words)

# 새로운 열에 추출된 단어 저장
df['extracted_word'] = df['inis'].apply(extract_word)

# 결과 출력
print(df[['inis', 'extracted_word']])
```

```python
data = df['inis']
df = pd.DataFrame(data)

# 제외할 문자 및 패턴 정의
exclude_patterns = r'공공기관|주식회사|상품권|유한회사|이니시스|\(.*?\)|[_]'

# 단어 추출 함수 정의
def extract_word(text):
    text = re.sub(exclude_patterns, '', text)  # 제외할 문자 제거
    text = re.sub(r'\b\d+\b', '', text)  # 독립된 숫자만 제거
    text = re.sub(r'\b[A-Za-z]\b', '', text)  # 독립된 영어 문자 제거
    text = re.sub(r'\b주식\w{0,2}\b', '', text)  # "주식"으로 시작하며 4글자인 단어 제거

    # 한글, 영어 및 숫자가 결합된 단어 추출
    words = re.findall(r'[가-힣A-Za-z0-9]+', text)

    # 중복 단어 제거(순서 유지)
    unique_words = list(dict.fromkeys(words))
    return ' '.join(unique_words)

# 새로운 열에 추출된 단어 저장
df['extracted_word'] = df['inis'].apply(extract_word)

# 'extracted_word' 열이 빈 경우 해당 행 삭제
df = df[df['extracted_word'].str.strip() != '']  # 공백 제거 후 빈 값인지 확인

output_file = "output.xlsx"
df.to_excel(output_file, index=False)

print(f"추출된 단어가 저장된 엑셀 파일로 내보냈습니다: {output_file}")
```
